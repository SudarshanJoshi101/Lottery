```python
import pandas as pd
import numpy as np
import random

# Runtime inputs
start_num = int(input("Enter starting number (inclusive): "))
end_num = int(input("Enter ending number (exclusive): "))
num_columns = int(input("Enter number of columns: "))
total_numbers = int(input("Enter total random numbers to generate: "))

# Validate range
if end_num - start_num < total_numbers:
    raise ValueError("Range too small for the number of unique values requested.")

# Step 1: Generate sorted random numbers
random_numbers = sorted(random.sample(range(start_num, end_num), total_numbers))

# Step 2: Pad if necessary to fill the table evenly
rows = (total_numbers + num_columns - 1) // num_columns
random_numbers += [None] * (rows * num_columns - total_numbers)

# Step 3: Convert to column-wise arrangement
array = np.array(random_numbers).reshape((rows, num_columns), order='F')  # column-wise fill
df = pd.DataFrame(array)

# Step 4: Set column headers to 1-based indexing
df.columns = [str(i) for i in range(1, num_columns + 1)]

# Step 5: Save to Excel
df.to_excel("random_number_table_sorted.xlsx", index=False)
print("âœ… File saved as 'random_number_table_sorted.xlsx'")
df.head()
```
